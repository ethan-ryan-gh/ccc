```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

class Gaus:
    def __init__(self, x, y, amplitude, mean, sigma, height):
        '''
        1. This init statement initilzes the inputs that are required for the gaussian function (guess parameters).
        2. It then sorts the guesses into a list to use for scipy's curve_fit class.
        3. The curve_fit class creates a list of optimized parameters and the covariance matrix by running through
            the singGaus function containing the equation of a Gaussian distribution.
        4. Finally the optimized parameters are fit to the gaussian function.
        
        To call this class simply use: var = Gaus(x={x_data}, y={y_data}, amplitude=, mean=, sigma=, height=)
        '''
        self.x = x
        self.y = y
        self.a = amplitude
        self.m = mean
        self.s = sigma
        self.h = height

        # Fit the Gaussian to the data
        param_guess = [self.a, self.m, self.s, self.h]
        self.popt, self.pcov = curve_fit(self.singGaus, self.x, self.y, p0=param_guess)
        
        # Generate the fitted Gaussian curve using the optimized parameters
        self.fit = self.singGaus(self.x, *self.popt)

    @staticmethod
    def singGaus(x, amplitude, mean, sigma, height):
        return amplitude * np.exp(-((x - mean)**2) / (2 * sigma**2)) + height

    def get_params(self):
        # Return the optimized parameters with uncertainties from covariance matrix
        uncertainties = np.sqrt(np.diag(self.pcov))
        print(f'Amplitude: {self.popt[0]} ± {uncertainties[0]}')
        print(f'Mean: {self.popt[1]} ± {uncertainties[1]}')
        print(f'Sigma: {self.popt[2]} ± {uncertainties[2]}')
        print(f'Height: {self.popt[3]} ± {uncertainties[3]}')

    def simpPlot(self, x_label, y_label):
        # Plots a simple diagram of the data that was passed through and the Gaussian fit.
        # To use: {var}.simpPlot(x_label={x_axis_label}, y_label={y_axis_label})
        plt.figure(figsize=(15, 10))
        plt.plot(self.x, self.y, 'o', label="Data")
        plt.plot(self.x, self.fit, 'r-', label="Fitted Gaussian")
        plt.xlabel(x_label)
        plt.ylabel(y_label)
        plt.legend()
        plt.show()

```

# Test

```python
import pandas as pd

##############
# Import data
##############

df_test = pd.read_csv("./H2D2/Hg_Spectra_546.csv") # create pandas dataframe
x_test = df_test["Wavelength"] # grab column
y_test = df_test["I"] # grab column


########################
# Input Known Constants
########################

a_Hg = 0.12 # amplitude                   
mean_Hg = 545.491 # middle of graph                 
h_Hg = 0.410 # vertical shift                  
sigma_Hg = 0.004 # width displacement


#############
# Test Class
#############

# Call Class
gaussian_fit = Gaus(x=x_test, y=y_test, amplitude=a_Hg, mean=mean_Hg, sigma=sigma_Hg, height=h_Hg)

# Print the optimized parameters with uncertainties
gaussian_fit.get_params()

# Plot the data and the fitted Gaussian
gaussian_fit.simpPlot(x_label='x', y_label='y')
```
